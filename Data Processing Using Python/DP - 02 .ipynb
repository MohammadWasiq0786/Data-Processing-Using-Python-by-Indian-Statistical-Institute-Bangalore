### FEATURE SELECTION
**Recursive Feature Elimination Method**
* A methodology to select most relevant features in a dataset .
* A way to select the important factors or fields in a dataset based on its usefulness in predicting a target or response variable .
* An effective way to eliminate the features not useful for predicting the response variable

**Recursive Feature Elimination Method**
1. Fit a model with all features
2. Validate the model
3. Calculate accuracy or score
4. Remove the least contributing feature
5. Fit the model with remaining feature and go back to step The procedure stops when all the remaining features are important.  <br>Generally accuracy is estimated using cross validation

**Recursive Feature Elimination Method - R**
* Continuous   <br>Commonly Used Models 
 1. Linear regression  
 2. General additive models
 3. Random Forest
 4. Bagging, etc.
* Categorical <br>Commonly Used Models 
 1. Logistic regression
 2. Linear discriminant analysis
 3. General additive models
 4. Random Forest
 5. Bagging, etc
 
**Recursive Feature Elimination Method - Python
<br>Example:-** Select the 5 important features in the Boston housing data. The response variable is MEDV.

```python
# Import packages
import pandas as pd
from sklearn.feature_selection import RFE
from sklearn import tree
# Import dataset
mydata = pd.read_excel("C://Users//ACER//Downloads//DP-02//Data/Boston_Housing Data.xls")
mydata.head()
```

```python
# Separate x's & y
x = mydata.iloc[:, 0:13]
y = mydata.MEDV
```

```python
# Set up the model
mymodel = tree.DecisionTreeRegressor() ; mymodel
```

**Recursive Feature Elimination Method - Python
<br>Example:-** Select the 5 important features in the Boston housing data. The response variable is MEDV.

```python
# Set up Recursive feature Elimination
mysearch = RFE(mymodel, n_features_to_select=5, step=1)

# Fit RFE to the data
mysearch = mysearch.fit(x,y)

# identify important features
decision = mysearch.support_  
print(decision)

decision = pd.DataFrame(decision, columns=["Decision"])
print(decision)

# featue ranking
rank = mysearch.ranking_
rank = pd.DataFrame(rank, columns=["Rank"])  
print(rank)
```

**Recursive Feature Elimination Method - Python
<br>Example:-** Select the 5 important features in the Boston housing data. The response variable is MEDV.

```python
# getting the feature names
features = x.columns
features = pd.DataFrame(features, columns= ["Features"])

# Preparing the result
myresult = features.join(decision)
myresult = myresult.join(rank)
myresult
```

```python
# Exporting the result
myresult.to_excel("C://Users//ACER//Downloads//DP-02/rfe_result.xlsx")
```

# Correlation & Regression
1. **Correlation :**
<br>Correlation analysis is a technique to identify the relationship between two variables.
<br>Type and degree of relationship between two variables.

**Usage -**
* Explore the relationship between the output characteristic and input or process variable.
* *Output variable* : **y** : Dependent variable
* *Input / Process variable* : **x** : Independent variable
* *Positive Correlation:* y increases as x increases & vice versa
* *Negative Correlation:* y decreases as x increases & vice versa
* *No Correlation:*

**Measure of Correlation:** 
* **Coefficient of Correlation**
<br>Correlation Coefficient $r = \frac{S_{xy}}{\sqrt{(S_{xx}.S_{yy})}}$ , 
<br>where , $S_{xy} = \sum(x-\bar{x})(y-\bar{y})$
<br>$S_{xx} = \sum(x-\bar{x})^2$
<br>$S_{yy} = \sum(y-\bar{y})^2$
<br>Symbol : **r**
<br>Range : **-1 to 1**
<br>Sign : **Type of correlation**
<br>Value : **Degree of correlation**
<br>If *correlation is positive*
<br>$S_{xy}$ will be **positive**
<br>If *correlation is negative*
<br>$S_{xy}$ will be **negative**

Examples:
<br>*r = 0.6* , 60 % positive correlation
<br>*r = -0.82* , 82% negative correlation
<br>*r = 0* , No correlation

## Multiple Regression Analysis
1. **Regression**
<br>Correlation helps
<br>To check whether two variables are related
<br>If related
<br>Identify the type & degree of relationship

Regression helps
* To identify the exact form of the relationship
* To model output in terms of input or process variables

**Simple Linear Regression**
* **General Form** : $y = a + bx + \epsilon$
<br>where , $a :$ Intercept (the value of y when x is equal to 0)
<br>$b :$ slope (Indicates the amount of change in y with every unit change in x)
* **Parameter Estimation**
$$y = a + bx + \epsilon$$
$\hat{a} = \bar{y} - \hat{b}\bar{x}$
<br>$\hat{b} = \frac{S_{xy}}{S_{xx}}$

<br>where, $S_{xy} = \sum(x-\bar{x})(y-\bar{y})$
<br>$S_{xx} = \sum(x-\bar{x})^2$

* **Test for Significance** 
<br>(Testing b = 0 or not) of relation between x & y
<br>$H_0 : b = 0$
<br>$H_1 : b \neq 0$

* Test Statistics
$$t_0 = \frac{(\hat{b} -0)}{se(\hat{b})}$$

* **Result** 
<br>If $p-value < 0.05$ , then $H_0$ is rejected & y can be modeled with x

* **Issues**
<br>For any set of data,
<br>a & b can be calculated
<br>Regression model $y = a + bx + \epsilon$ can be build
<br>*But all the models may not be useful*

* **Coefficient of Regression:**
<br>Measure of degree of Relationship
<br>$R^2$
<br>$R^2 = \frac{SS_R}{S_{yy}} = b.\frac{S_{xy}}{S_{yy}}$
<br>$SS_R = \sum(y_{predicted} - \bar{y})^2$
<br>$S_{yy} = \sum(y_{actual} - \bar{y})^2$
<br>$R^2 :$ Amount variation in y explained by x 
<br>**Range of** $R^2 :$ **0** to **1**
<br> If $R^2 \geq 0.6$ , the model is reasonable good .

**Multiple Linear Regression**
<br>To model output variable y in terms of two or more variables.
* **General Form:**
<br>$y = a + b_1x_1 + b_2x_2 + - - - + b_kx_k + \epsilon$
<br>Two variable case:
<br>$y = a + b_1x_1 + b_2x_2 + \epsilon$
<br>Where , $a:$ intercept (the predicted value of y when all xs are zero)
<br>$b_j:$ slope (the amount change in y for unit change in $x_j$ keeping all other xs constant, j = 1,2,---,k)

## Regression with Dummy variables
When xs are not numeric but nominal
<br>Each nominal or categorical variable is converted into dummy variables
<br>Dummy variables takes values 0 or 1
<br>Number of dummy variable for one x variable is equal to number of distinct
<br>values of that variable - 1

**Example:** A study was conducted by an IT company to develop a model to estimate sprint productivity of agile projects in telecom vertical. The explonatory variables chosen are developer skill, review type and code reuse. Data was collected from 34 projects and is given in Agile_Productivity file.?

```python
# Load Libraries
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
import seaborn as sns

# Load Data
mydata = pd.read_excel("C://Users//ACER//Downloads//DP-02/Data/Agile_Sprint_Productivity.xlsx")
print(mydata.head())

dev_skill = mydata.Developer_Skill
rev_skill = mydata.Review_Type
reuse = mydata.iloc[:,2]
sp = mydata.Sprint_Productivty
```

```python
# Regression with dummy variables
# Checking relation between x and y
# Developer Skill Vs Sprint Productivity

sns.boxplot(x = dev_skill, y = sp)
plt.title("Developer Skill Vs Sprint Productivity")
plt.xlabel("Developer Skill")
plt.ylabel("Sprint Productivity")
plt.show()
```

```python
# Review Type Vs Sprint Productivity
sns.boxplot(x = rev_skill, y = sp)
plt.title("Review Type Vs Sprint Productivity")
plt.xlabel("Review Type")
plt.ylabel("Sprint Productivity")
plt.show()
```

```python
# Reuse Vs Sprint Productivity
sns.scatterplot(x = reuse, y = sp)
plt.title("Reuse Vs Sprint Productivity")
plt.xlabel("Reuse")
plt.ylabel("Sprint Productivity")
plt.show()
```

```python
# Regression with dummy variables  Output
mymodel = ols("sp ~ dev_skill + rev_skill + reuse" , mydata).fit()
mymodel.summary()
```

```python
# Regression with dummy variables  Normality test of Residuals
predy = mymodel.predict()
res = sp - predy
stats.probplot(res, plot= myplot)
myplot.grid()
myplot.show()
```

```python
# Regression with dummy variables  Normality test of Residuals
mytest = stats.normaltest(res)
test_stat = round(mytest[0],4)
test_stat
p_value = round(mytest[1],4)
p_value

print("w =" , test_stat)

print("p_value =" , p_value)
```

# BINARY LOGISTIC REGRESSION
Used to develop models when the output or response variable y is binary
<br>The output variable will be binary, coded as either success or failure
<br>Models probability of success p which lies between 0 and 1
<br>Linear model is not appropriate
$$p =\frac{e^{a+b_1x_1 + a+b_2x_2 + \cdots + a+b_kx_k}}{1 + e^{a+b_1x_1 + a+b_2x_2 + \cdots + a+b_kx_k}}$$
where , $p:$ probability of success
<br>$x_is :$ independent variables
<br>$a, b_1 , b_2, ---:$ coefficients to be estimated
<br>If estimate of $p \geq 0.5$ , then classified as $success$, otherwise as $failure$

**Example:** Develop a model to predict the non payment of overdrafts by customers of a multinational banking institution. The data collected is given in Logistic_Reg.csv file. The factors and response considered are given below.

<br>2 --- Transaction speed score
<br>3 --- Peer comparison score in terms of transaction volume

Response --- Values
<br>Outcome --- 0: Not Paid and 1: Paid
